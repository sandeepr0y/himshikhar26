# 9. Optimization Example

Optimization is the process of finding the best value that minimizes error or loss.
In Deep Learning, optimization helps neural networks learn the best weights for accurate predictions.

Minimize:f(x)=x^2



**Gradient Descent:** Gradient Descent **minimizes a function** step-by-step.
It **continuously updates the value of x** to move toward the minimum point of the function.

```python
x = 10 # initial point some random values
learning_rate = .2  # how big each update step is

for i in range(20):

    gradient = 2*x  # Compute Gradient by computing derivative.

    x -= learning_rate * gradient   # This is the Gradient Descent update rule:  x_new = x_old −η × gradient

    print(f"Iteration {i}: x = {x}")
```

```python
import matplotlib.pyplot as plt

x = 10
learning_rate = .1  # how big each update step is

_x = []
_y = []

for i in range(20):
    gradient = 2*x  # Compute Gradient by computing derivative.
    x -= learning_rate * gradient   # This is the Gradient Descent update rule:  x_new = x_old −η × gradient
    _x.append(x)
    _y.append(x**2)

plt.plot(_x, _y, marker='x')

plt.title("y = x^2")
plt.xlabel("x")
plt.ylabel("y")

plt.show()
```

## 19. Mean Squared Error (MSE)
Mean Squared Error (MSE) measures the **average squared difference between actual and predicted values**.
It is commonly used in Deep Learning and **regression problems to measure prediction error.**

MSE.png

```python
import numpy as np

actual = np.array([10, 20, 30])
predicted = np.array([12, 18, 33])

mse = np.mean((actual - predicted)**2)  # MSE formula implemented
print(mse)

MSE = lambda prediction, ground_truth: np.mean((ground_truth - prediction)**2)
print(f"MSE function: ", MSE(predicted, actual))
```

- Errors are squared
- Large errors get more penalty
- Smaller MSE means better model performance

## 22. Gradient Descent Weight Update

Gradient Descent updates the weights of a neural network to reduce prediction error or loss.
It helps the model learn the best weights for making accurate predictions during training.

**Formula**

Wnew = Wold −η × gradient

**Where:**

- W_old  = current weight
- η = learning rate
- gradient = slope of loss function

**Example**

**Suppose:**

weight = 5
learning rate = 0.1
gradient = 2

```python
weight = 5
learning_rate = 0.1
gradient = 2

new_weight = weight - learning_rate * gradient

print(new_weight)
```

The weight changes from 5 to 4.8

Gradient Descent moves weights in the direction that reduces loss

Repeating this process helps neural networks learn better predictions

## 25. Create Neural Network in PyTorch
A neural network in PyTorch is **created using layers and activation functions to learn patterns from data.**
PyTorch provides the **nn.Module class** to easily **build** **custom neural network** architectures.

```python
import torch
import torch.nn as nn  # Neural Network

class NeuralNetwork(nn.Module):     # nn.Module (every NN inherit from it)

    def __init__(self):            # __init__ Constructor Function initializes: layers, variables, network structure

        super(NeuralNetwork, self).__init__()      # initializes: parent class

        self.fc1 = nn.Linear(2, 4)     # 1st FC layer  2: input 4 is output
        self.relu = nn.ReLU()          # Relu Activation function
        self.fc2 = nn.Linear(4, 1)     # 2nd FC layer    4: input , 1: output

    def forward(self, x):              # data moves through network

        x = self.fc1(x)                # learn features from data x
        x = self.relu(x)               # introduces nonlinearity
        x = self.fc2(x)                # generates prediction

        return x
```

- fc1 is the first fully connected layer
- ReLU introduces nonlinearity
- fc2 is the output layer
- forward() defines how data moves through the network

This **neural network learns patterns by updating weights during training**.

## Create Model

## Training Loop
A training loop **repeatedly performs forward propagation**, **loss computation**, **backpropagation**, and w**eight updates** to **train the neural network.**
It **helps** the model **gradually reduce error** and **improve prediction accuracy over multiple epochs**.

```python
for epoch in range(10):

    prediction = model(x)

    target = torch.tensor([[1.0]])

    loss = criterion(prediction, target)

    optimizer.zero_grad()

    loss.backward()  # pytorch automatically computes gradients,  applies chain rule, updates computation graph.

    optimizer.step()

    print(f"Epoch {epoch}: Loss = {loss.item()}")
```

- **prediction = model(x)** performs forward propagation
- **loss = criterion(prediction, target)** computes error
- **loss.backward()** computes gradients using backpropagation
- **optimizer.step()** updates weights
The loop repeats for multiple epochs to improve learning.

As training continues:

- loss decreases
- predictions become more accurate
- neural network learns patterns from data

## Convolutional Neural Network (CNN)
A Convolutional Neural Network (CNN) is a Deep Learning model specially designed for **image processing** and **computer vision tasks**.
**CNNs** automatically **learn** important **image features** such as **edges**, **textures**, **shapes**, and **objects**.

```python
class CNN(nn.Module):

    def __init__(self):

        super(CNN, self).__init__()

        self.conv1 = nn.Conv2d(1, 16, 3)
        self.relu = nn.ReLU()
        self.pool = nn.MaxPool2d(2)

    def forward(self, x):

        x = self.conv1(x)
        x = self.relu(x)
        x = self.pool(x)

        return x
```

 - **Conv2d** performs convolution operation on images
ReLU introduces nonlinearity
- **MaxPool2d** reduces image size and extracts important features

**CNNs are widely used in:**

- image classification
- face recognition
- object detection
- medical image analysis
- biometric recognition

CNNs learn image patterns automatically without manual feature extraction.

```python
model.load_state_dict(torch.load("model.pth"))
```

- **torch.load()** loads the saved model file
- **load_state_dict()** restores learned weights into the neural network
- **model.eval()** sets the model to evaluation mode

**Loading models is useful for:**

- testing on new data
- deployment
- inference
- real-world AI applications

```python
######################
# Select device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Move model to device
model.to(device)

# Create input tensor
test_input = torch.tensor([[1.0, 2.0]])

# Move input to same device
test_input = test_input.to(device)

# Evaluation mode
model.eval()

# Prediction
with torch.no_grad():

    prediction = model(test_input)

print(prediction)
```
